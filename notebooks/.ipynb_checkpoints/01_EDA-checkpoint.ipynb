```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# show plots inline
%matplotlib inline

# Load the dataset
df = pd.read_csv("../data/diabetes.csv")
df.shape, df.columns.tolist()
```

```python
df.head()
df.info()
df.describe().T
```

# Check class balance

```python
print(df['Outcome'].value_counts())
sns.countplot(x='Outcome', data=df)
plt.title("Outcome class balance (0: No Diabetes, 1: Diabetes)")
plt.show()
```

```python
likely_missing = ['Glucose','BloodPressure','SkinThickness','Insulin','BMI']
for col in likely_missing:
    zeros = (df[col] == 0).sum()
    print(f"{col}: {zeros} zeros ({zeros/len(df)*100:.1f}%)")
```

# Distributions and relationships

```python
df.hist(figsize=(12,10), bins=20)
plt.tight_layout()
plt.show()

# Correlation heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Feature Correlation Matrix")
plt.show()
```

# Boxplots by Outcome

```python
plt.figure(figsize=(14,10))
for i, col in enumerate(df.columns[:-1], 1):
    plt.subplot(3,3,i)
    sns.boxplot(x='Outcome', y=col, data=df)
    plt.title(col)
plt.tight_layout()
plt.show()
```

# FINDINGS


 Dataset has 768 records, 9 columns.
 Class balance  65% no-diabetes / 35% diabetes.
 Zero values found in Glucose, BloodPressure, BMI, Insulin  will impute.
 Glucose, BMI, Age show strong difference between classes.
